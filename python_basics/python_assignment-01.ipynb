9:Type Casting:Type casting is used to change the datatype of a variable in python.

```python
example of typecasting
my_age=input("Enter your age:") #before typecasting
print(type(my_age))

my_age=int(input("Enter your age")) #after typecasting
print(type(my_age))

weight=56.9
print(type(weight)) #before typecasting
print(type((int(weight)))) #after typecasting
```

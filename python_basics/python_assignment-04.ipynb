## **Example#01:Remove pairs untill condition fails**

```python
s = "aabbcc33"
i=0
while True:
    if len(s)>=2 and s[:2]==s[0]*2: #e.g:"aa"=="aa" then condition will be True
        s=s[2:]
    elif  len(s)>=1 and s[-1].isdigit(): #check if a character is digit
        continue
    else:
        print("s:",s)
        print("String end!")
        break
        
```

## **Example#02:Indexing + continue filter**

```python
s = "school2026"
new_s=""
for i in range(0,len(s)):
    if s[i]=="a" or s[i]=="e" or s[i]=="i" or s[i]=="o" or s[i]=="u": #checks if a character is vowel
        continue
    elif s[i].isdigit(): #check if a character is digit
        break
    else:
        new_s=new_s+s[i] #appending values in second string
print(new_s)
```

## **Example#03:Reverse Window Scanner**

```python
s = "helloo@@"
count=0
while True:
    if len(s)>=3 and s[-2:]==s[-1]*2: #e.g:"@@"=="@@" then condition will be True
        s=s[:-2] #remove last two characters from the string
        count+=1
    elif not s[-1].isalnum(): #alnum() check whether a character is an alphabet or number
        continue
    else:
        break
print(s)
print(count)

```
